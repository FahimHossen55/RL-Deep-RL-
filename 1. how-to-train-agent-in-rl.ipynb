```python
import gym 
```

```python
import numpy as np  
clifEnv =  gym.make("CliffWalking-v0")
```

```python
done = False 
state =  clifEnv.reset() # starting a env 
```

```python
while not done: 

    print(clifEnv.render(mode = "ansi"))
    action = int(np.random.randint(low =0 , high = 4 , size = 1))
    print(state, "--->" , action)
    state, reward , done , _ = clifEnv.step(action)


clifEnv.close()
```
