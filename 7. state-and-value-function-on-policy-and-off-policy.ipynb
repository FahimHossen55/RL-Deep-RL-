#  OFF policy and on policy  



## ON policy: 
### In on-policy RL, the agent learns and improves the same policy it uses to interact with the environment. The policy being evaluated and optimized is identical to the policy generating the actions (behavior policy). 

- The behavior policy (for exploration) = the target policy (being learned).
- Typically, the policy includes some randomness (e.g., ε-greedy or softmax) to ensure exploration.
- Data used for learning comes from the current policy, so older data from previous policies becomes obsolete as the policy improves.

# Off-Policy RL  
## In off-policy RL, the agent learns a target policy (the policy being optimized) that is different from the behavior policy used to interact with the environment. The behavior policy generates data, while the target policy is learned from that data. 

- Behavior policy (for exploration) ≠ target policy (being learned).
- Often uses a replay buffer to store experiences (states, actions, rewards, next states) from various policies, allowing reuse of old data.
- The behavior policy is typically more exploratory (e.g., random or ε-greedy) to cover the state-action space, while the target policy may be deterministic or greedy.
