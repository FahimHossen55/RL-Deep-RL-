# **Actor Critic**  
Actor- Critic (AC) is type of reinforcement leanring algorithm that combines elements of both policy based methods (Actor) and value based methods ( Critic) to learn a poicy that can maximize the expected cumulative reward.  
<br> 
The Actor - Critic alg,orithm has two main components : 
- actor
-  Critic

The actor- critic algorithm combines the strenght oif both value based and policy based methods. The critic provides the vaue functions estimates, ehich help to reduce the variance of the policy updates while the actor provides the policy updates that are necessary . 

# **A3C** 
## A3C stands for Asychronous Advantage Actor-Critic. It is deep reinforcement learning algorithm that  combines the advantages ob both actor critic and deep Q learning. 
<br>  

The key idea behind A3C is to use multiple agents each running in parallel and learning from their own experiences , to speed up the learning process.  
<br> 

In A3C, there are two main components : 
- An actor network
- And a critic network

The actor network is responsible for learning the policy or the mapping from states  to actions , The critic network on the other hand learns tha value function or the expected return for a given state.
